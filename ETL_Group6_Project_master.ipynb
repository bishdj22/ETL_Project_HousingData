```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import os
from citipy import citipy
```

```python
csv_path = "CleanData/single_fam_data_master2016-18.csv"
single_fam_data = pd.read_csv(csv_path, encoding='ISO-8859-1',dtype={'zip': 'str'})
single_fam_data['zip'] = single_fam_data['zip'].apply(lambda x: x.zfill(5))
single_fam_data.head(50)
```

```python
zips_analyzed = single_fam_data['zip'].nunique()
cities_analyzed = single_fam_data['city'].nunique()
metros_analyzed = single_fam_data['metro'].nunique()
counties_analyzed = single_fam_data['countyname'].nunique()
banks_analyzed = single_fam_data['bank'].nunique()
records_analyzed = single_fam_data['zip'].count()
filter_ltv_above = single_fam_data[single_fam_data['ltv'] >= 0.8]
filter_ltv_below = single_fam_data[single_fam_data['ltv'] <= 0.8]
ltv_above = filter_ltv_above['ltv'].count()
ltv_below = filter_ltv_below['ltv'].count()

geosumm_dict = {"Records":records_analyzed,
                "Zip Codes":zips_analyzed, 
                "Cities": cities_analyzed, 
                "Metropolitan Areas":metros_analyzed,
                "Counties":counties_analyzed,
                "Banks":banks_analyzed,
                "ltv above 80%":ltv_above,
                "ltv below 80%":ltv_below}
                

geo_coverage_df = pd.DataFrame.from_dict(geosumm_dict, orient='index', columns=['Total'])
geo_coverage_df
```
