```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
zillow_2016_df=pd.read_csv('/Users/maheshdivakarannair/Documents/GitHub/ETL_Project_HousingData/CleanData/Zillow_2016avghomeval.csv')
```

```python
zillow_2016_df.head()
```

```python
rds_connection_string = "mahesh1:mahesh@localhost:5432/housing_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
zillow_2016_df.to_sql(name='zillow_db', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select count(*) from zillow_db', con=engine).head()
```
