```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import os
from citipy import citipy
```

```python
csv_2016_path = "CleanData/Zillow_2016avghomeval.csv"
csv_2018_path = "CleanData/Zillow_2018avghomeval.csv"
```

```python
data_2016_df = pd.read_csv(csv_2016_path)
data_2018_df = pd.read_csv(csv_2018_path)
```

```python
data_2016_df.head()
```

```python
data_2018_df.head()
```

```python
join_data_df = pd.merge(data_2016_df, data_2018_df, how='outer', on='ZipCode')
join_data_df = join_data_df.drop(["City_y","State_y","Metro_y","CountyName_y"], axis=1).rename(columns={"City_x":"City",\
                                                                                               "State_x":"State",\
                                                                                               "Metro_x":"Metro",\
                                                                                               "CountyName_x":"County"})

join_data_df.head(10)
```

```python
join_data_df.to_csv("CleanData/Zillow_2016_2018_avghomeval.csv",index=False)
```

```python
NY_data_df = join_data_df.loc[(join_data_df["State"]=="NY")]
NY_data_df.head(10)
```

```python
highest_city_df = join_data_df.sort_values('2016_AvgHomeValue', ascending=False).drop_duplicates(['City']).reset_index()

highest_city_df = highest_city_df.sort_values('State', ascending=False).dropna().reset_index()
highest_city_df
```
